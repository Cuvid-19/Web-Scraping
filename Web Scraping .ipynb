```python
# bs4 / BeautifulSoup = library for screen-scraping
from bs4 import BeautifulSoup

#request= get https library
import requests
```

```python
url = 'https://www.coingecko.com'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
```

```python
response = requests.get(url, headers=headers)
#check link status!
print(response)
```

```python
#get request to html
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
#Make code readable / clean
print(soup.prettify())
```

```python
#find/ filter using table / class name
soup.find('table', class_='tw-border-y tw-border-gray-200 dark:tw-border-moon-700 tw-divide-y tw-divide-gray-200 dark:tw-divide-moon-700 tw-w-full sortable')
```

```python
tables = soup.find_all('table')
print(tables)
```

```python
# Filter the table 
world_titles = [table.find_all('th') for table in tables ]
world_titles
```

```python
# Flatten the list of <th> elements
world_titles_flat = [th for table_titles in world_titles for th in table_titles]

# Extract text content and remove leading/trailing whitespace
world_table_titles = [title.text.strip() for title in world_titles_flat]

# Print the result
print(world_table_titles)
```

```python
#pandas
import pandas as pd
df =pd.DataFrame(columns = world_table_titles)
df
```

# Pulling and combine data


```python
column_data = [table.find_all('tr') for table in tables ]

#read row by row and strip 
for rows in column_data:
    for row in rows[1:]:
        row_data = row.find_all('td')
        individual_row_data = [data.text.strip() for data in row_data]
        print(individual_row_data)
        
        length = len(df)
        df.loc[length] = individual_row_data
```

```python
df
```

```python
# SORT the rows of dataframe by '24h Volume' column 
df2 = df.sort_values(by = '24h Volume') 
  
# show the resultant Dataframe 
df2 
```
